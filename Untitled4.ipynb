```python
import json
import pandas as pd
import re
import matplotlib.pyplot as plt
from collections import Counter
from datetime import datetime, timedelta
import seaborn as sns
from wordcloud import WordCloud
json_files = ['backend.json', 'devops.json', 'front.json', 'fullstack.json']
dfs = [pd.read_json(file) for file in json_files]
combined_df = pd.concat(dfs, ignore_index=True)
```

```python
print(len(combined_df))
```

```python
combined_df['title'] =combined_df['title'].str.lower()

front_end = combined_df[combined_df['title'].str.contains('front')]
print(len(front_end))
```

```python
full_stack = combined_df[combined_df['title'].str.contains('stack')]
print(len(full_stack))
```

```python
devops = combined_df[combined_df['title'].str.contains('devops')]
print(len(devops))
```

```python
backend = combined_df[combined_df['title'].str.contains('back')]
print(len(backend))
```

```python
soft_jobs = pd.concat([front_end, backend, devops,full_stack], ignore_index=True)
print(len(soft_jobs))
print(soft_jobs.head())
print(soft_jobs.isnull().sum())
def convert_posted_date(text):
    try:
        text = text.strip().replace('Posted', '').replace('ago', '').strip()

        if 'day' in text:
            days = int(text.replace('days', '').replace('day', '').strip())
            posted_date = datetime.today() - timedelta(days=days)
        elif 'month' in text:
            months = int(text.replace('months', '').replace('month', '').strip())
            posted_date = datetime.today() - timedelta(days=months * 30)
        else:
            posted_date = pd.NaT
        return posted_date.date()

    except Exception as e:
        print(f"Error: {e}")
        return pd.NaT

# تطبيق الدالة
soft_jobs['posted_date'] = soft_jobs['posted_date'].apply(convert_posted_date)
print(soft_jobs['posted_date'].head())
```

```python
job_counts = soft_jobs['title'].value_counts()

plt.figure(figsize=(30,10))  # كبرت عرض الرسمه جدا
bars = plt.bar(job_counts.index, job_counts.values, color='skyblue', width=0.5)

plt.title('Distribution of Software Job Titles', fontsize=22)
plt.xlabel('Job Title', fontsize=18)
plt.ylabel('Number of Postings', fontsize=18)
plt.xticks(rotation=75, ha='right', fontsize=12)

# كتابة الأرقام فوق الأعمدة
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, int(yval), ha='center', va='bottom', fontsize=10)

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
print('Top 20 Software Job Titles')
top_jobs = job_counts.head(20)  # عرض أول 20 وظيفة بس

plt.figure(figsize=(18,8))
bars = plt.bar(top_jobs.index, top_jobs.values, color='skyblue', width=0.5)  # عرض أعمدة أعرض شوية

plt.title('Top 20 Software Job Titles', fontsize=18)
plt.xlabel('Job Title', fontsize=14)
plt.ylabel('Number of Postings', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)

# كتابة الأرقام فوق الأعمدة
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, int(yval), ha='center', va='bottom', fontsize=10)

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
all_skills = []

for skills in combined_df['skills']:
    if isinstance(skills, str):  # إذا كان skills نص مكتوب، مش NaN
        skill_list = [skill.strip().lower() for skill in skills.split(',')]
        all_skills.extend(skill_list)

# -----------------------------
# Count top skills
# -----------------------------
skill_counts = Counter(all_skills)
top_10_skills = skill_counts.most_common(20)

# Convert to DataFrame for plotting
top_skills_df = pd.DataFrame(top_10_skills, columns=['Skill', 'Count'])

# -----------------------------
# Plot Visualization
# -----------------------------
plt.figure(figsize=(12, 6))
plt.barh(top_skills_df['Skill'], top_skills_df['Count'], color='skyblue')
plt.xlabel('Number of Jobs')
plt.ylabel('Skill')
plt.title('Top 20 Software Engineering Skills')
plt.gca().invert_yaxis()  # عشان يجيب أعلى skill فوق
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
json_files_sci = ['dataAnalyst.json', 'dataEng.json', 'dataScient.json']
dfs_sci = [pd.read_json(file) for file in json_files_sci]
combined_df_sci = pd.concat(dfs_sci, ignore_index=True)
combined_df_sci['title'] = combined_df_sci['title'].str.lower()

print(len(combined_df_sci))
print(combined_df_sci.head())
```

```python
# Sample text data (you can replace this with your data)
text = ' '.join(combined_df_sci['title'].astype(str))

# Create a WordCloud object
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(text)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")  # Turn off axis numbers and ticks
plt.show()
```

```python
# List of keywords/phrases to search for
keywords = ['analyst','data entry','analytics','data analysis','data scient','support engineer','data engineer','data architect','senior engineer']

# Convert the list of keywords into a regular expression pattern (case-insensitive)
pattern = '|'.join(keywords)

# Filter the DataFrame for titles that contain any of the keywords (case-insensitive)
filtered_df = combined_df_sci[combined_df_sci['title'].str.contains(pattern, case=False, na=False)]
print(len(filtered_df))
```

```python
print(filtered_df.isnull().sum())
```

```python
# 1. Split the skills string into a list
combined_df_sci['skills'] = combined_df_sci['skills'].str.split(',')

# 2. Flatten all the skills into one big list
all_skills = combined_df_sci['skills'].explode()

# 3. Remove extra spaces from each skill (like ' Python' -> 'Python')
all_skills = all_skills.str.strip()

# 4. Count the frequency of each skill
skill_counts = all_skills.value_counts()

# 5. Show the top 20 most frequent skills
print(skill_counts.head(20))

# Get top 20 skills
top_skills = skill_counts.head(20)

# Normalize the counts for coloring
norm = plt.Normalize(top_skills.min(), top_skills.max())
colors = plt.cm.viridis(norm(top_skills))  # You can change 'viridis' to another colormap

# Create figure and axis manually
fig, ax = plt.subplots(figsize=(12,6))

# Now plot on the given ax
bars = ax.barh(top_skills.index, top_skills.values, color=colors)

# Add numbers on bars
for bar in bars:
    ax.text(bar.get_width() + 1,
            bar.get_y() + bar.get_height()/2,
            int(bar.get_width()),
            va='center')

# Create colorbar
sm = plt.cm.ScalarMappable(cmap='viridis', norm=norm)
sm.set_array([])  # required for colorbar
cbar = fig.colorbar(sm, ax=ax)  # <--- attach to the correct ax
cbar.set_label('Number of Job Listings Requiring Skill')

# Labels and style
ax.set_xlabel('Number of Job Listings Requiring Skill')
ax.set_title('Top 20 Most Required Skills')
ax.invert_yaxis()
ax.grid(axis='x')

plt.show()
```

```python
def count_job_types(df, dataset_name):
    hybrid = 0
    onsite = 0
    remote = 0

    columns_to_search = ['description', 'title']

    # لو الداتا فيها عمود full_text ضيفه
    if 'full_text' in df.columns:
        columns_to_search.append('full_text')

    for col in columns_to_search:
        for text in df[col].dropna():
            text = text.lower()
            if 'hybrid' in text:
                hybrid += 1
            if 'on site' in text:
                onsite += 1
            if 'remote' in text:
                remote += 1

    return pd.Series({'Hybrid': hybrid, 'On Site': onsite, 'Remote': remote}, name=dataset_name)

# ------------------------------------------
# Apply function on all datasets
# ------------------------------------------
sci_counts = count_job_types(filtered_df, 'filtered_df')
cyb_counts = count_job_types(combined_df_cyb, 'combined_df_cyb')
soft_counts = count_job_types(soft_jobs, 'soft_jobs')

# Combine all results
all_counts = sci_counts + cyb_counts + soft_counts

# ------------------------------------------
# Plot Pie Chart
# ------------------------------------------
plt.figure(figsize=(7, 7))
plt.pie(all_counts, labels=all_counts.index, autopct='%1.1f%%', startangle=100, colors=['#66b3ff','#99ff99','#ffcc99'])
plt.title('Job Types Distribution (Hybrid, On Site, Remote)')
plt.axis('equal')  # عشان الدايرة تبقى مظبوطة
plt.show()
```

```python
def count_job_types(df, dataset_name):
    hybrid = 0
    onsite = 0
    remote = 0

    columns_to_search = ['description']
    for col in columns_to_search:
        for text in df[col].dropna():
            text = text.lower()
            if 'hybrid' in text:
                hybrid += 1
            if 'on site' in text:
                onsite += 1
            if 'remote' in text:
                remote += 1

    return pd.Series({'Hybrid': hybrid, 'On Site': onsite, 'Remote': remote}, name=dataset_name)

# ------------------------------------------
# Apply function on all datasets
# ------------------------------------------
sci_counts = count_job_types(filtered_df, 'filtered_df')
cyb_counts = count_job_types(combined_df_cyb, 'combined_df_cyb')
soft_counts = count_job_types(soft_jobs, 'soft_jobs')

# Combine all results
all_counts = sci_counts + cyb_counts + soft_counts

# ------------------------------------------
# Plot Pie Chart
# ------------------------------------------
plt.figure(figsize=(7, 7))
plt.pie(all_counts,
        labels=all_counts.index,
        autopct='%1.1f%%',
        startangle=100,
        colors=['#66b3ff','#99ff99','#ffcc99'],
        labeldistance=1.1,  # Move the labels further out for better visibility
        textprops={'fontsize': 14}  # Increase font size of the labels
       )
plt.title('Job Types Distribution (Hybrid, On Site, Remote)', fontsize=16)  # Increase title font size
plt.axis('equal')  # Make the chart a perfect circle
plt.show()
```
